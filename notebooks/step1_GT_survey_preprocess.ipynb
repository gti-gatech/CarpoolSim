## Load GT survey data

```python
gt_survey = pd.read_csv(
    os.path.join(
        "..", "data_inputs", "gt_survey",
        "2022_campus_commute_survey_data_no_zip_na_05122023.csv"
    ),
    index_col = 0
)

print(gt_survey.shape)
```

```python
# filter out nan values for zip code
gt_survey = gt_survey[gt_survey['ZIPCODE'].notna()]
gt_survey["ZIPCODE"] = gt_survey["ZIPCODE"].astype(int)

display(gt_survey.head(3))
print(gt_survey.shape)
```

### filter dataset using survey answers

```python
gt_survey.Q2.unique()
```

```python
# Q10: home to work time
# Q11: work to home time
gt_survey.Q10.unique()
```

```python
filt = (gt_survey.Q10.notna())
gt_survey = gt_survey[gt_survey.Q10.notna()]
print(gt_survey.shape)
```

```python
gt_survey['Q10'] = gt_survey['Q10'].str.replace(
    "12:00am - 5:30am",
    "5:30am"
)
```

```python
# 2023-01-01 is the pseudo date
depart_time = "2023-01-01 " + gt_survey.Q10

gt_survey['depart_time'] = pd.to_datetime(
    depart_time,
    format=r"%Y-%m-%d %I:%M%p"
)
```

```python
def extract_newmins(df_row):
    tm = (df_row["depart_time"] - pd.to_datetime("2023-01-01")
         ).total_seconds() / 60
    return tm

gt_survey['newmin'] = gt_survey['depart_time'] - pd.to_datetime("2023-01-01")
gt_survey['newmin'] = gt_survey.apply(extract_newmins, axis=1)
gt_survey.head()
```

```python
# package for querying zip code
import pgeocode

locator = pgeocode.Nominatim("us")
location = locator.query_postal_code(30310)
print((location.latitude, location.longitude))
```

```python
# extract longitude/latitude for all positions!
def extract_lon_lat(df_row):
    location = locator.query_postal_code(df_row["ZIPCODE"])
    return pd.Series({
        "longitude": location.longitude,
        "latitude": location.latitude
    })

gt_survey[["ori_lon", "ori_lat"]] = gt_survey.apply(extract_lon_lat, axis=1)
display(gt_survey.head(3))
```

```python
from shapely.geometry import Point

# create a point geometry
gt_survey["geometry"] = [
    Point(xy) for xy in zip(gt_survey.ori_lon, gt_survey.ori_lat)
]
```

## Assign TAZs to centroids

```python
# load TAZ file
pth = os.path.join(rt, "data_inputs/ABM2020 203K/taz/TAZ_combine.shp")
gdf = gpd.read_file(pth)
gdf.head()
```

```python
# assign TAZ given long/lat
def get_taz_given_lon_lat(df_row, tazs=gdf):
    point = df_row["geometry"]
    taz_geoms = tazs["geometry"].tolist()
    filt = [taz_geom.contains(point) for taz_geom in taz_geoms]
    
    if sum(filt) == 0:
        the_taz = -1
    else:
        the_taz = tazs.loc[filt, "TAZS"].iloc[0]
    return the_taz

gt_survey["orig_taz"] = gt_survey.apply(get_taz_given_lon_lat, axis=1)
```

```python
# add Geogia Tech as final destination
gt_survey["dest_taz"] = 484
gt_survey["dest_lon"] = -84.397971
gt_survey["dest_lat"] = 33.775766
```

```python
filt = (gt_survey["orig_taz"] != -1)
gt_survey = gt_survey.loc[filt, :]
print(gt_survey.shape)
```

```python
# load zone
import pickle

pickle_pth = os.path.join(
    os.environ['PROJ_LIB'],
    "data_inputs",
    'taz_zones.pickle'
)
with open(pickle_pth, 'rb') as handle:
    taz_zone = pickle.load(handle)

taz2region = {}
for region in taz_zone:
    lst = taz_zone[region]
    for taz in lst:
        taz2region[int(taz)] = region
# print(taz2region)
def map_taz2zone(taz):
    return taz2region[int(taz)]

gt_survey['O_region'] = gt_survey.orig_taz.apply(map_taz2zone)
gt_survey['D_region'] = gt_survey.dest_taz.apply(map_taz2zone)
```
